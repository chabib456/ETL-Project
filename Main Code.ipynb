```python

parking_id = []
address = []
rate = []
with open("Resources/greenPParking2015.json") as json_file:  
    data = json.load(json_file)
    for p in data['carparks']:
        parking_id.append(p["id"])
        address.append(p["address"])
        rate.append(p["rate"])
parking_df=pd.DataFrame({"parking_id":parking_id,
                        "address":address,
                        "rate":rate})
parking_df.head()
```
